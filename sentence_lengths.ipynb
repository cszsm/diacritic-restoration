```python
import pandas as pd
%matplotlib inline
import matplotlib

import src.corpus_counter as counter
```

```python
ndigits = -2

lengths = counter.get_sentence_lengths()
rounded_lengths = [round(length, ndigits) for length in lengths]

df_rounded_lengths = pd.DataFrame(rounded_lengths)
```

```python
df_rounded_lengths.hist(0, figsize=(15, 10))
```

```python
filtered_rounded_lengths = [length for length in rounded_lengths if length <= 1000]

df_filtered_rounded_lengths = pd.DataFrame(filtered_rounded_lengths)
```

```python
df_filtered_rounded_lengths.hist(0, figsize=(15, 10))
```
